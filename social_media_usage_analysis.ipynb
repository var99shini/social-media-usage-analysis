```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("social_media_usage.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
app_usage = df.groupby("App")["Time_Spent_Min"].sum()

plt.figure()
app_usage.plot(kind="bar")
plt.title("Total Time Spent on Each App")
plt.xlabel("Social Media App")
plt.ylabel("Time Spent (Minutes)")
plt.show()
```

```python
day_usage = df.groupby("Day")["Time_Spent_Min"].sum()

plt.figure()
day_usage.plot(kind="line", marker='o')
plt.title("Daily Social Media Usage Trend")
plt.xlabel("Day")
plt.ylabel("Time Spent (Minutes)")
plt.show()
```

```python
purpose_usage = df.groupby("Purpose")["Time_Spent_Min"].sum()

plt.figure()
purpose_usage.plot(kind="pie", autopct='%1.1f%%')
plt.title("Purpose of Social Media Usage")
plt.ylabel("")
plt.show()
```

##  Key Insights from Social Media Usage Analysis

- Instagram accounts for the highest total usage time among all apps.
- Social media usage is higher during weekends compared to weekdays.
- Entertainment is the dominant purpose of social media usage.
- YouTube shows significant usage for learning and educational purposes.
